JSON:JavaScript object notation is a format used to store and transfer data, which is usually used to transfer data from server to web page. JSON is a lightweight text data exchange format, similar to XML, but smaller, faster, easier to understand and parse than XML.

Json.loads ():The function can decode the data and transform the JSON data into Python data type, so that it can process the JSON data in Python. Use json.loads () function, passing JSON data into parentheses can convert string type data into dictionary type data.
